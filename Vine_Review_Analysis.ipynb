```python
import pandas as pd
```

```python
# read and check dataset
vine_df = pd.read_csv("Resources/vine_table.csv")
vine_df.head(10)
```

```python
# filter df to keep values
# specified total_votes count >= 20
vine_filter_df = vine_df[vine_df['total_votes'] >= 20]
vine_filter_df.head(10)
```

```python
# add restrictions to find helpful reviews
formula = vine_filter_df['helpful_votes']/vine_filter_df['total_votes'] >= 0.5
vine_helpful_df = vine_filter_df[formula]
vine_helpful_df.head(10)
```

```python
# review was written as part of the Vine program (paid)
review_df = vine_helpful_df[vine_helpful_df['vine'] == 'Y']
review_df.head(10)
```

```python
# review was not part of the Vine program (unpaid)
nonreview_df = vine_helpful_df[vine_helpful_df['vine'] == 'N']
nonreview_df.head(10)
```

```python
# calculate number of total reviews
# for paied and unpaid vine
total_review = review_df['review_id'].count()
total_nonreview = nonreview_df['review_id'].count()

print(f'number of reviews for paid vine: {total_review}')
print(f'number of reviews for unpaid vine: {total_nonreview}')
```

```python
# calculate number of total 5star-reviews
# for paied and unpaid vine
review5 = review_df[review_df['star_rating'] == 5].count()['review_id']
nonreview5 = nonreview_df[nonreview_df['star_rating'] == 5].count()['review_id']

print(f'number of 5 star reviews for paid vine: {review5}')
print(f'number of 5 star reviews for unpaid vine: {nonreview5}')
```

```python
# calculate percentages of total 5star-reviews
# for paied and unpaid vine
review5_percent = review5/total_review * 100
nonreview5_percent = nonreview5/total_nonreview * 100

print(f'Percentage of 5 star review for paid vine: {review5_percent:.2f}%')
print(f'Percentage of 5 star review for unpaid vine: {nonreview5_percent:.2f}%')
```
